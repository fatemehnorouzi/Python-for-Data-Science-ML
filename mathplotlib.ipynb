```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.linspace(0, 5, 11)
y= x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
fig= plt.figure()
axes= fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig= plt.figure(figsize= (6, 3), dpi=100)
axes1= fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2= fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes2.plot(y, x, 'r')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Larger plot')
axes2.set_title('Smaller plot')
```

```python
fig, axes= plt.subplots(nrows= 1, ncols= 2, figsize= (4, 3), dpi=100)
plt.tight_layout()
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'r')
```

```python
fig.savefig('my_pic.jpg')
```

```python
fig= plt.figure(figsize= (6, 3), dpi=100)
axes= fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, x**2, label= 'Squared', color= 'purple', linewidth=2, alpha=0.5, ls='--', marker='o')
axes.plot(x, x**3, label= 'Cubed' , linewidth=2, alpha=0.5, ls='--', marker='o')
axes.legend(loc=0)
```

```python
axes.set_xlim([0,1])
axes.set_xlim([0,2])
```
